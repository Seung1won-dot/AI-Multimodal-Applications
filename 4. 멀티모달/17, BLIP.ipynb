# **1. BLIP**
BLIP(Bootstrapping Language-Image Pretraining)는 이미지와 텍스트를 함께 학습하는 Vision-Language 모델로, 이미지 캡셔닝과 이미지-텍스트 매칭을 동시에 수행할 수 있도록 설계된 멀티모달 사전학습 모델입니다. 이미지 인코더와 텍스트 인코더를 결합해 이미지와 문장을 같은 표현 공간에 정렬하며, 노이즈가 포함된 웹 데이터에서도 자동 캡션 생성과 필터링을 통해 학습 데이터를 정제하는 “bootstrapping” 전략을 사용합니다. 이를 통해 이미지 설명 생성(Image Captioning), 텍스트-이미지 검색(Retrieval), VQA 등 다양한 멀티모달 작업에서 강력한 성능을 보이는 것이 특징입니다. [[논문]](https://arxiv.org/pdf/2201.12086)

### 1. Bootstrapping의 의미
Bootstrapping Language-Image Pretraining은 Bootstrapping이 핵심입니다. 웹에서 수집한 이미지-텍스트 데이터는 캡션이 부정확하거나 이미지와 맞지 않거나 노이즈가 많습니다. BLIP은 이를 해결하기 위해 모델이 스스로 캡션을 생성하고, 그 중 좋은 데이터를 선별해 다시 학습에 사용하는 전략을 도입했습니다. 즉, 데이터를 정제하면서 스스로 학습을 개선하는 구조입니다.

1. 초기 모델로 캡션 생성
2. 원래 캡션과 비교
3. 품질 좋은 샘플만 선별
4. 선별된 데이터로 다시 학습

# **2. BLIP의 전체 구조**

① Vision Encoder
이미지를 feature vector로 변환합니다.

- ViT (Vision Transformer) 기반
이미지를 패치로 나눠 Transformer로 처리
- 출력: 이미지 임베딩

② Text Encoder

텍스트를 임베딩 벡터로 변환합니다.

- BERT 스타일 Transformer
- 텍스트를 토큰 단위로 인코딩
- 출력: 텍스트 임베딩

③ Multimodal Encoder / Decoder

- 이미지 특징 + 텍스트 특징을 함께 처리
- Cross-Attention을 통해 두 모달리티 결합

> 이 부분이 이미지-텍스트 매칭와 이미지 캡션 생성을 가능하게 합니다.

# **2. BLIP-2**
BLIP-2는 이미지와 대형 언어모델(LLM)을 효율적으로 연결하기 위해 설계된 Vision-Language 모델로, 기존 BLIP과 달리 이미지 인코더와 LLM을 직접 미세조정하지 않고, 그 사이에 Q-Former(Querying Transformer)라는 중간 모듈을 두어 두 모델을 연결하는 구조를 사용합니다. 이미지에서 추출한 시각 특징을 Q-Former가 소수의 학습 가능한 쿼리 벡터로 압축하고, 이를 LLM이 이해할 수 있는 형태로 변환하여 텍스트 생성에 활용합니다. 이 방식은 대규모 LLM을 그대로 유지하면서도 이미지 정보를 효과적으로 주입할 수 있어, 이미지 캡셔닝·VQA·멀티모달 질의응답 등에서 강력한 성능을 보이면서도 학습 비용을 크게 줄인 것이 핵심 특징입니다. [[논문]](https://arxiv.org/pdf/2301.12597)

### 1. BLIP2의 배경
기존 Vision-Language 모델들의 문제점은 다음과 같습니다.

- 이미지 인코더 + 텍스트 인코더 + 멀티모달 인코더를 모두 새로 학습
- 학습 데이터 수억~수십억 개 필요
연산량이 매우 큼
- LLM을 직접 멀티모달로 다시 학습해야 함

> 👉 BLIP-2는 “이미 잘 학습된 모델은 그대로 두고, 연결만 잘하자”라는 전략을 사용합니다.

### 2.  BLIP-2의 핵심 아이디어

Pretrained Vision Encoder (고정)
    - 예: ViT, CLIP Vision Encoder
    - Freeze (가중치 고정)

Pretrained LLM (고정)
    - 예: OPT, Flan-T5
    - Freeze (가중치 고정)

Q-Former (Query Transformer)
    - 이미지 정보를 LLM이 이해할 수 있도록 중간에서 번역하는 역할
    - 이것만 학습함

### Q-Former
- Learnable Query 벡터를 여러 개 둠 (예: 32개)
- 이 Query들이 이미지 feature와 cross-attention 수행
- 이미지에서 중요한 정보만 추출
- LLM에 전달할 작은 feature 집합 생성

> “이미지에서 LLM이 이해할 수 있는 핵심 정보만 뽑아주는 압축기”

# **3. BLIP-2 학습 단계**

### 1. Vision-Language Representation Learning
- Vision Encoder + Q-Former 학습
- Image-Text Matching
- Image Captioning

👉 이 단계에서 Q-Former가 이미지-텍스트 정렬을 학습합니다.

### 2. LLM 연결 학습
- Q-Former 출력 → LLM 입력으로 연결
- LLM은 고정
- Q-Former만 학습

> 👉 결과적으로 LLM이 이미지를 이해하는 것처럼 보이게 됩니다.

# **4. BLIP과 CLIP으로 텍스트에서 이미지 검색 성능을 비교하는 예제**

```python
import torch
import numpy as np

from PIL import Image
from datasets import load_dataset

from transformers import BlipProcessor, BlipForImageTextRetrieval
from transformers import CLIPProcessor, CLIPModel
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
# BLIP
blip_processor = BlipProcessor.from_pretrained('Salesforce/blip-itm-base-coco')
blip_model = BlipForImageTextRetrieval.from_pretrained('Salesforce/blip-itm-base-coco').to(device)
blip_model.eval()

# CLIP
clip_model = CLIPModel.from_pretrained('openai/clip-vit-base-patch32').to(device)
clip_processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')
clip_model.eval()
```

```python
# wds_flickr8k 데이터셋의 txt 컬럼의 타입을 str 변환
def normalize_caption(x):
    if x is None:
        return ""
    if isinstance(x, (bytes, bytearray)):
        return x.decode('utf-8', errors='ignore')
    if isinstance(x, dict):
        for k in ['text', 'caption', 'txt', 'value']:
            if k in x and isinstance(x[k], str):
                return x[k]
        return str(x)
    if isinstance(x, (list, tuple)):
        for item in x:
            if isinstance(item, str):
                return item
        return str(x)
    if isinstance(x, str):
        return x
    return str(x)
```

```python
# wds_flickr8k 데이터셋의 jpg 컬럼의 타입을 모두 PIL로 통일
def to_pil(x):
    if isinstance(x, Image.Image):
        return x.convert('RGB')
    if isinstance(x, dict) and 'bytes' in x and x['bytes'] is not None:
        import io
        return Image.open(io.BytesIO(x['bytes'])).convert('RGB')
    if isinstance(x, str):
        return Image.open(x).convert('RGB')
    raise TypeError(f'지원하지 않는 이미지 형식: {type(x)}')
```

```python
def l2_normalize(x, axis = -1, eps = 1e-12):
    # keepdims: L2 norm: 유클리드 거리
    # L2 정규화 후 내적하면 코사인 유사도가 됨
    return x / (np.linalg.norm(x, axis=axis, keepdims=True) + eps)
```

```python
def evaluate_retrieval(similarity_matrix, top_k=(1, 5, 10)):
    num_samples = similarity_matrix.shape[0]
    ranks = []
    for i in range(num_samples):
        # i번째 텍스트에 대해 모든 이미지 점수를 정렬(내림차순)
        sorted_idx = np.argsort(similarity_matrix[i])[::-1]
        # 정렬된 인덱스 중에서 정답 이미지 i가 몇 번째 위치인지 확인
        # rank_pos = 0 이면 1등 (Top-1)
        rank_pos = int(np.where(sorted_idx == i)[0][0])
        ranks.append(rank_pos)
    results = {f'Top-{k}': float(np.mean(np.array(ranks) < k)) for k in top_k}
    return results
```

```python
# assert: 코드 내에서 디버깅 및 검증하는 데 사용되는 도구
# True일 것을 가정하고 코드를 작성. 조건이 False일 경우 AssertionError 발생
def get_blip_scores(images_pil, texts, batch_size=16, use_itm_head=True):
    assert len(images_pil) == len(texts)
    n = len(images_pil)
    scores = np.zeros((n, n), dtype=np.float32)

    with torch.no_grad():
        for i in range(n):
            # 같은 이미지 i를 n번 복제해서 배치로 저장
            img_batch = [images_pil[i]] * n
            # 한 번의 forward에서 (image_i, text_0), (image_i, text_1), ... (image_i, test_(n-1))
            txt_batch = texts
            inputs = blip_processor(images=img_batch, text=txt_batch, padding=True, return_tensors='pt').to(device)
            if use_itm_head:
                # ITM 헤드를 사용해서 점수를 받아옴
                # 버전에 따라 필드명이 다를 수 있어 아래와 같이 처리
                out = blip_model(**inputs, use_itm_head=True)
                if hasattr(out, 'itm_score') and out.itm_score is not None:
                    itm = out.itm_score
                elif hasattr(out, 'logits') and out.logits is not None:
                    itm = out.logits
                else:
                    itm = out[0]

                # itm 결과가 (batch, 2)면 두 개의 클래스(불일치/일치) 로짓으로 저장됨
                # positive(일치) 로밋인 [:, 1]만 뽑아서 '매칭 점수'로 사용할 수 있음
                # 만약 shape가 다르면 1차원으로 펼쳐서 사용
                if itm.dim() == 2 and itm.size(-1) == 2:
                    row = itm[:, 1]
                else:
                    row = itm.view(-1)
            else: # itm head를 쓰지 않은 경우
                out = blip_model(**inputs, use_itm_head=False)
                row = out[0].view(-1)
            scores[i, :] = row.detach().float().cpu().numpy()
    return scores
```

```python
def get_clip_embeddings(images_pil, texts, batch_size=32):
    image_embeds_list = []
    text_embeds_list = []

    with torch.no_grad():
        for s in range(0, len(images_pil), batch_size):
            batch_imgs = images_pil[s:s + batch_size]
            inputs = clip_processor(images=batch_imgs, return_tensors='pt', padding=True).to(device)
            vision_out = clip_model.vision_model(pixel_values=inputs['pixel_values'])
            img_feat = clip_model.visual_projection(vision_out.pooler_output)
            image_embeds_list.append(img_feat.detach().cpu().numpy())

        for s in range(0, len(texts), batch_size):
            batch_txt = texts[s:s + batch_size]
            inputs = clip_processor(text=batch_txt, return_tensors='pt', padding=True, truncation=True).to(device)
            text_out = clip_model.text_model(input_ids=inputs['input_ids'], attention_mask=inputs['attention_mask'])
            txt_feat = clip_model.text_projection(text_out.pooler_output)
            text_embeds_list.append(txt_feat.detach().cpu().numpy())

    image_embeds = np.concatenate(image_embeds_list, axis=0)
    text_embeds = np.concatenate(text_embeds_list, axis=0)
    return image_embeds, text_embeds
```

```python
dataset = load_dataset('clip-benchmark/wds_flickr8k')['test']
sample_size = 100
subset = dataset.shuffle(seed=2026).select(range(sample_size))

raw_images = list(subset['jpg'])
raw_cpations = list(subset['txt'])

images = [to_pil(x) for x in raw_images]
captions = [normalize_caption(x) for x in raw_cpations]

keep = [i for i, c in enumerate(captions) if len(c.strip()) > 0]
images = [images[i] for i in keep]
captions = [captions[i] for i in keep]

n = len(images)
print('final sample size: ', n)
```

```python
# BLIP
blip_similarity_matrix = get_blip_scores(images, captions, use_itm_head=True)
blip_results = evaluate_retrieval(blip_similarity_matrix)
for k, v in blip_results.items():
    print(f'{k}: {v:.4f}')
```

```python
clip_image_embeds, clip_text_embeds = get_clip_embeddings(images, captions)
clip_image_embeds = l2_normalize(clip_image_embeds)
clip_text_embeds = l2_normalize(clip_text_embeds)
clip_similarity_matrix = clip_image_embeds @ clip_text_embeds.T
clip_results = evaluate_retrieval(clip_similarity_matrix)
for k, v in clip_results.items():
    print(f'{k}: {v:.4f}')
```
