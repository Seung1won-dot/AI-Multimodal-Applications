# 1. Multimodal Learning
Multimodal Learning은 서로 다른 여러 모달리티의 데이터를 함께 학습하는 방식입니다. 예를 들어 이미지와 텍스트를 동시에 입력으로 받아 장면을 설명하거나, 영상과 음성을 함께 활용해 감정을 인식하는 모델이 이에 해당합니다. 각 모달리티가 가진 상호보완적인 정보를 결합함으로써 단일 모달리티보다 더 풍부한 표현과 높은 성능을 기대할 수 있습니다. 최근에는 비전과 언어를 결합한 모델들이 대표적인 성공 사례로 주목받고 있습니다.

### 💬 Unimodal model

Unimodal model은 하나의 데이터 모달리티만을 입력으로 사용하는 모델입니다. 예를 들어 텍스트만을 사용하는 감성 분류 모델, 이미지 데이터만을 입력으로 하는 이미지 분류 모델, 음성 신호만을 분석하는 음성 인식 모델 등이 이에 해당합니다. 입력과 출력이 단일 형태로 고정되어 있기 때문에 구조가 비교적 단순하고 학습과 해석이 용이하며, 특정 데이터 유형에 최적화된 성능을 내는 데 강점이 있습니다. 다만 현실 세계의 복잡한 문제처럼 여러 정보원이 동시에 필요한 상황에서는 표현력에 한계가 있습니다.

### 💬 Multitask Learning

Multitask Learning은 하나의 모델이 여러 개의 서로 다른 작업(task)을 동시에 학습하도록 설계하는 방법입니다. 예를 들어 하나의 신경망이 문장에 대해 감성 분류와 주제 분류를 함께 수행하거나, 객체 탐지와 객체 분류를 동시에 학습하는 경우가 이에 해당합니다. 여러 작업이 공통으로 사용하는 표현을 공유함으로써 일반화 성능이 향상되고, 데이터가 부족한 작업에서도 학습 안정성을 얻을 수 있다는 장점이 있습니다. 이는 모델이 보다 보편적인 특징을 학습하도록 유도하는 전략이라고 볼 수 있습니다.

# 2. 멀티모달 모델의 종류

### 1. Vision–Language 모델 (이미지 + 텍스트)
이미지와 텍스트를 함께 이해하거나 서로 변환하는 모델입니다. 이미지를 보고 설명을 생성하거나(Image Captioning), 이미지에 대한 질문에 답하는(Visual Question Answering), 텍스트로 이미지를 검색하는(Text-to-Image Retrieval) 작업에 사용됩니다. 현재 멀티모달 연구와 서비스에서 가장 널리 사용되는 형태입니다.

### 2. Audio–Visual 모델 (음성 + 영상)
음성과 영상 정보를 동시에 활용하는 모델입니다. 입 모양과 음성을 함께 사용한 음성 인식, 얼굴 표정과 음성 톤을 결합한 감정 인식, 영상 속 인물의 행동 분석 등에 사용됩니다. 소음 환경에서도 강건한 인식이 가능하다는 장점이 있습니다.

### 3. Audio–Language 모델 (음성 + 텍스트)
음성을 텍스트와 함께 처리하는 모델입니다. 음성 인식(STT), 음성 기반 질의응답, 음성 명령 이해 등에 활용되며, 대화형 AI나 음성 비서 시스템의 핵심 구성 요소입니다.

### 4. Vision–Audio–Language 모델 (이미지/영상 + 음성 + 텍스트)
이미지(또는 영상), 음성, 텍스트를 모두 입력으로 받아 종합적으로 이해하는 모델입니다. 영상 요약, 멀티모달 질의응답, 상황 이해 기반 추론 등 복잡한 작업을 수행할 수 있으며, 최근 대규모 멀티모달 파운데이션 모델에서 주로 채택되는 구조입니다.

### 5. Sensor Fusion 기반 멀티모달 모델 (비전 + 센서)
카메라, 라이다, 레이더, GPS 등 다양한 센서 데이터를 결합하는 멀티모달 모델입니다. 자율주행, 로보틱스, 스마트 팩토리 분야에서 사용되며, 환경 인식과 의사결정의 정확도를 높이는 것이 핵심 목적입니다.

### 6. Document Multimodal 모델 (텍스트 + 이미지 + 표)
문서 내의 텍스트, 이미지, 표, 레이아웃 정보를 함께 처리하는 모델입니다. OCR, 문서 요약, 문서 질의응답, 계약서·보고서 자동 분석 등에 활용됩니다.

# 3. Multimodal Fusion
멀티모달 모델에서는 각 모달리티가 형태·차원·의미 표현 방식이 모두 다르기 때문에, 이를 그대로 섞을 수 없습니다. 따라서 각 모달의 정보를 적절한 시점과 수준에서 결합(fusion)하여 하나의 의미 표현으로 만드는 과정이 필요하며, 이 결합 전략을 Multimodal Fusion 방식이라고 합니다.

### 1. Early Fusion (초기 결합)
Early Fusion은 입력 단계에서 모달리티를 바로 결합하는 방식입니다. 예를 들어 이미지 특징 벡터와 텍스트 임베딩을 추출한 뒤 이를 하나의 벡터로 단순 연결(concatenation)하여 모델에 입력합니다. 구조가 단순하고 구현이 쉬우며, 모달 간 상호작용을 초반부터 학습할 수 있다는 장점이 있습니다. 다만 모달리티 간 노이즈가 그대로 섞일 수 있고, 한 모달이 다른 모달을 방해하는 문제가 발생할 수 있습니다.



### 2. Late Fusion (후기 결합)
Late Fusion은 각 모달리티를 독립적으로 처리한 뒤, 최종 결과를 결합하는 방식입니다. 예를 들어 이미지 분류 모델과 텍스트 분류 모델을 각각 학습한 뒤, 두 결과를 평균이나 가중합으로 합쳐 최종 예측을 수행합니다. 모달별 모델을 독립적으로 최적화할 수 있어 안정성이 높고, 한 모달이 없어도 동작하기 쉽다는 장점이 있습니다. 반면 모달 간의 깊은 상호작용을 학습하기는 어렵습니다.

### 3. Intermediate Fusion (중간 결합, Feature-level Fusion)
Intermediate Fusion은 각 모달리티에서 의미 있는 특징을 추출한 뒤, 중간 은닉 표현 단계에서 결합하는 방식입니다. CNN이나 Transformer로 각 모달의 특징을 뽑은 뒤, 어텐션이나 MLP를 이용해 결합하는 구조가 대표적입니다. Early Fusion보다 안정적이고 Late Fusion보다 표현력이 뛰어나며, 최근 멀티모달 모델에서 가장 널리 사용되는 방식입니다.

### 💬 Attention-based Fusion (어텐션 기반 결합)
Attention-based Fusion은 어떤 모달의 정보가 더 중요한지를 학습적으로 선택하는 방식입니다. 예를 들어 텍스트 질문을 기준으로 이미지의 특정 영역에 집중하거나, 음성 신호 중 감정이 강한 구간에 가중치를 부여할 수 있습니다. Cross-Attention, Co-Attention 구조가 대표적이며, VQA나 멀티모달 추론 모델에서 핵심적인 역할을 합니다. [중간 단계(feature-level) 결합 방법을 제안한 논문](https://arxiv.org/pdf/2107.00135)

### 5. Hybrid Fusion (혼합 결합)
Hybrid Fusion은 Early, Intermediate, Late Fusion을 조합한 방식입니다. 예를 들어 중간 단계에서는 어텐션 기반 결합을 사용하고, 최종 출력 단계에서는 결과를 다시 결합하는 구조입니다. 대규모 멀티모달 파운데이션 모델에서 자주 사용되며, 성능은 뛰어나지만 구조가 복잡하고 계산 비용이 큽니다.



# 4. Pretrained Vision Language Models (VLMs)
Pretrained Vision Language Models는 이미지와 텍스트를 동시에 이해하도록 설계된 멀티모달 모델을 대규모 이미지–텍스트 쌍 데이터로 미리 학습(pretraining)한 모델입니다. 사전학습 단계에서 이미지의 시각적 특징과 텍스트의 의미 표현을 같은 임베딩 공간에서 정렬(alignment)하도록 학습하기 때문에, 이후에는 이미지 캡셔닝, 이미지–텍스트 검색, 시각적 질의응답(VQA), 멀티모달 추론 등 다양한 작업에 소량의 추가 학습이나 미세조정만으로도 효과적으로 활용할 수 있습니다. 즉, Vision과 Language 사이의 공통 의미 표현을 미리 학습해두어 여러 멀티모달 태스크에 범용적으로 사용할 수 있는 기반 모델이라고 볼 수 있습니다.

### 1. CLIP 계열 (Contrastive Learning 기반)
CLIP은 이미지와 텍스트를 각각 인코더로 처리한 뒤, 두 임베딩을 같은 공간에서 대비 학습(contrastive learning)하는 모델입니다. 이미지–텍스트 검색(Text-to-Image, Image-to-Text)에 매우 강하며, zero-shot 분류가 가능해 Vision–Language 사전학습 모델의 출발점으로 자주 언급됩니다.



### 2. ALIGN 계열 (대규모 웹 데이터 기반)
ALIGN은 매우 큰 규모의 이미지–텍스트 쌍 데이터를 활용해 사전학습된 모델로, CLIP과 유사한 대비 학습 구조를 가지지만 데이터 스케일을 극단적으로 확장한 것이 특징입니다. “데이터 규모가 성능을 만든다”는 관점을 잘 보여주는 모델입니다.



### 3. ViLBERT / LXMERT 계열 (Dual-stream Transformer)
ViLBERT, LXMERT는 이미지와 텍스트를 각각 Transformer로 처리한 뒤 Cross-Attention으로 상호작용시키는 dual-stream 구조를 사용합니다. 이미지–텍스트 간의 관계 추론이 중요한 VQA, 멀티모달 추론 문제에서 많이 사용되었으며, 구조적으로 “멀티모달 어텐션”을 명확히 보여주는 교과서적인 모델입니다.

### 4. BLIP / BLIP-2 계열 (이해 + 생성 통합)
BLIP, BLIP-2는 이미지–텍스트 이해와 생성(Image Captioning, VQA)을 하나의 프레임워크로 통합한 모델입니다. 캡셔닝, QA, 검색을 모두 잘 수행하도록 설계되었으며, 최근 Vision–Language 모델의 실용적 활용을 대표하는 구조입니다.

### 5. Flamingo 계열 (Few-shot 멀티모달 LLM)
Flamingo는 대규모 언어모델(LLM)에 비전 인코더를 연결한 구조로, 소량의 예시(few-shot)만으로도 다양한 멀티모달 태스크를 수행할 수 있습니다. “Vision + LLM 결합” 흐름을 본격화한 대표적인 모델입니다.


### 6. 최신 멀티모달 파운데이션 모델 계열
최근에는 Vision–Language 모델이 단독 태스크용이 아니라, 범용 멀티모달 파운데이션 모델로 확장되고 있습니다. 이미지 이해, 문서 이해, 추론, 생성까지 하나의 모델로 처리하는 방향으로 발전하고 있으며, Vision–Language가 LLM 생태계의 핵심 구성 요소가 되고 있습니다.

### 💬  Zero-shot learning VS Few-shot learning
Zero-shot learning은 모델이 학습 과정에서 한 번도 보지 못한 클래스나 작업에 대해, 추가적인 예시 없이 바로 추론을 수행하는 방식입니다. 모델은 사전학습(pretraining) 단계에서 습득한 일반적인 언어·개념 지식이나 의미 공간(semantic space)을 활용해 입력을 해석합니다. 예를 들어, 이미지–텍스트 모델에서 “a photo of a zebra”라는 설명만으로 얼룩말 이미지를 찾아내는 경우가 이에 해당합니다. 이 방식의 핵심은 레이블 간의 의미적 관계를 이해하고 있다는 점이며, 새로운 작업에 대한 확장성과 유연성이 매우 높지만, 구체적인 도메인 특화 성능은 제한될 수 있습니다.

Few-shot learning은 새로운 작업이나 클래스에 대해 아주 적은 수의 예시(보통 1~10개 내외)만 제공한 뒤 그 패턴을 빠르게 일반화하도록 하는 방식입니다. 모델은 입력과 함께 주어진 소수의 예시를 컨텍스트로 활용해 문제의 구조를 파악하고 추론합니다. 예를 들어, 대규모 언어 모델에 질문과 함께 몇 개의 예시 Q&A를 제시하면, 그 형식과 규칙을 학습한 것처럼 동작합니다. Few-shot은 zero-shot보다 정확도와 안정성이 높아지는 경향이 있으나, 예시의 품질과 대표성에 성능이 크게 좌우됩니다.
