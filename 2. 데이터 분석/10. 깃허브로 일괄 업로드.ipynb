# GitHub로 일괄 업로드하기

이 튜토리얼에서는 Google Colab에서 작성한 파일들을 GitHub 저장소에 일괄 업로드하는 방법을 학습합니다.

## 1. GitHub Personal Access Token 생성

1. GitHub에 로그인합니다.
2. Settings → Developer settings → Personal access tokens → Tokens (classic)로 이동합니다.
3. Generate new token을 클릭합니다.
4. 필요한 권한을 부여합니다 (repo 항목에 체크).
5. 토큰을 복사합니다. (⚠️ 토큰은 다시 확인할 수 없으니 안전한 곳에 저장하세요)
